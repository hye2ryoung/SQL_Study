```python
import pandas as pd
tmp = pd.read_excel(stations_files[0], header=2)
tmp.tail(2)
```

```python
tmp_raw = []
for file_name in stations_files:
    # 엑셀 파일을 판다스 데이터 프레임 저장
    tmp = pd.read_excel(file_name, header = 2)
    # 데이터 프레임을 리스트 안에 저장
    tmp_raw.append(tmp)
```

```python
stations_raw = pd.concat(tmp_raw)
stations_raw
```

#### 의미 있는 데이터만 뽑아 사용하기

```python
stations_raw.head(1)
```

```python
data = {
    '상호': stations_raw['상호'],
    '주소': stations_raw['주소'],
    '휘발유': stations_raw['휘발유'],
    '경유': stations_raw['경유'],
    '셀프': stations_raw['셀프여부'],
    '상표': stations_raw['상표'],
}
```

```python
#raw데이터에서 의미 있는 데이터만 뽑아서 사용하기
stations = pd.DataFrame(data)
stations.tail()
```

```python
# 구 컬럼 추가
# 구 데이터를 리스트 형태로 저장
stations_gu_list = []
for address in stations['주소']:
    gu = address.split()[1]
    stations_gu_list.append(gu)

```

```python
stations_gu_list
```

```python
#데이터 검증 작업
len(stations_gu_list), len(stations)
```

```python
# 구 컬럼 추가
stations['구'] = stations_gu_list
```

```python
stations.head()
```

#### 가격 데이터 타입, 인덱스 재정렬

```python
stations['휘발유'].astype('float')
```

```python
#가격 정보 없는 주유소
stations[stations['휘발유'] == '-']
```

```python
#가격 정보가 있는 주유소만 사용
stations = stations[stations['휘발유'] != '-']
stations = stations[stations['경유'] != '-']
```

```python
# 데이터타입 object -> float 변환
stations['휘발유'] = stations['휘발유'].astype('float')
stations['경유'] = stations['경유'].astype('float')
```

```python
stations.info()
```

```python
#인덱스 재정렬
stations.reset_index(drop=True, inplace=True)
```
